## 1 зад: Кои са с най-положителните рейтинги, спрямо листа с всички популярни web страници?

## №1 За да проверим кои web страници са с най-позитивен рейтинг трябва да превърнем колоните 'Facebook_likes', 'Twitter_mentions', 'Google_pluses' от obj. във float тип.

```python
# it's calculating the number based on there abbriviation at the end
# example 19k = 19000; 1M = 1_000_000
def RewriteCol(col):
  for i, value in col.items():

    # if the value its count more than a million
    if(value.endswith('M')):
      col[i] = RenderNums(value,1_000_000)
      continue
    # if the value its count more than a thousand
    if(value.endswith('K')):
      col[i] = RenderNums(value,1_000)
      continue
    # if it's NaN value
    if(value == '-'):
      col[i] = 0
      continue

    # if the value doesn't meet the conditions
      col[i] = float(value)

  return col

def RenderNums(value, num):

      value = value[:-1]
      value = float(value)
      value = value * num
      return value

```

## Забележка: None стойностите в този data set не са нулеви стойности. Те отразяват не намерената информация за определен сайт за това "None" ще го заместим с 0.

```python
RewriteCol(web_data['Facebook_likes'])
web_data['Facebook_likes'] = web_data['Facebook_likes'].replace('None', 0).astype(float)

RewriteCol(web_data['Twitter_mentions'])
web_data['Twitter_mentions'] = web_data['Twitter_mentions'].replace('None', 0).astype(float)

RewriteCol(web_data['Google_pluses'])
web_data['Google_pluses'] = web_data['Google_pluses'].replace('None', 0).astype(float)
```

## №2 Чрез тази модификация можем да изчислим кои web страници са с най-положителните ревюта.

```python
#this function draws the histograms with positive reviews
def PositiveReviewsStats(websites,ratingScale):

    plt.hist(websites[ratingScale])
    plt.title('Based of ' + ratingScale)
    plt.xlabel('Reviews')
    plt.ylabel("Sites")
    plt.show()

#this function sort the values base of the given column and remove the duplicated values
def MostRatedSites(basedOf, index):
  return web_data.sort_values(basedOf,ascending=False).drop_duplicates(basedOf, keep='first').head(index)

Web_Index = 20 #How many rows do we want to extract
```

## Страниците с най-положително ревю спрямо: **Google**

```python
PositiveReviewsStats(MostRatedSites('Google_pluses',Web_Index), 'Google_pluses')
MostRatedSites('Google_pluses',Web_Index)
```

## Страниците с най-положително ревю спрямо: **Facebook**

```python
PositiveReviewsStats(MostRatedSites('Facebook_likes',Web_Index), 'Facebook_likes')
MostRatedSites('Facebook_likes',Web_Index)
```

## Страниците с най-положително ревю спрямо: **Twitter(компанията X)**

```python
PositiveReviewsStats(MostRatedSites('Twitter_mentions',Web_Index), 'Twitter_mentions')
MostRatedSites('Twitter_mentions',Web_Index)
```
